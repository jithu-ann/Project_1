```python
import pandas as pd 
from pathlib import Path
import matplotlib.pyplot as plt
import scipy.stats as st
import matplotlib.figure
```

```python
#Path to our csv files
stock_data = Path("Resources/ETF_prices.csv")

#Reading csv files into pandas
stock_df = pd.read_csv(stock_data)
stock_df
```

```python
#stock of fund symbol AAAU
stock_fund_df = stock_df[stock_df['fund_symbol']=='AAAU']
stock_fund_df
```

```python
#stock of the entire year of 2019,2020,2021
stock_all_2019_df = stock_df[stock_df['price_date'].between('2019-01-01','2019-12-31')]
stock_all_2020_df = stock_df[stock_df['price_date'].between('2020-01-01','2020-12-31')]
stock_all_2021_df = stock_df[stock_df['price_date'].between('2021-01-01','2021-12-31')]
stock_all_2021_df

```

```python
#2019 data of AAAU
stock_2019_df = stock_fund_df[stock_fund_df['price_date'].between('2019-01-01','2019-12-31')]
stock_2019_df
```

```python
#2020 data of AAAU
stock_2020_df = stock_fund_df[stock_fund_df['price_date'].between('2020-01-01','2020-12-31')]
stock_2020_df
```

```python
#2021 data of AAAU
stock_2021_df = stock_fund_df[stock_fund_df['price_date'].between('2021-01-01','2021-12-31')]
stock_2021_df
```

```python
#JAN 2019 data
jan_2019_df = stock_fund_df[stock_fund_df['price_date'].between('2019-01-01','2019-01-31')]
jan_2019_df
```

```python
#Plot of the closing price in jan 2019
stock_plot = plt.plot(jan_2019_df['price_date'],jan_2019_df['close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing price of 2019')
plt.xticks(rotation="vertical")
plt.show()
```

```python
#JAN 2020 data
jan_2020_df = stock_fund_df[stock_fund_df['price_date'].between('2020-01-01','2020-01-31')]
jan_2020_df
```

```python
#Plot of the closing price in jan 2020
stock_plot = plt.plot(jan_2020_df['price_date'],jan_2020_df['close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing price of 2020')
plt.xticks(rotation="vertical")
plt.show()
```

```python
#jan 2021 data
jan_2021_df = stock_fund_df[stock_fund_df['price_date'].between('2021-01-01','2021-01-31')]
jan_2021_df
```

```python
 #Plot of the closing price in jan 2021
stock_plot = plt.plot(jan_2021_df['price_date'],jan_2021_df['close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing price of 2021')
plt.xticks(rotation="vertical")
plt.show()
```

```python
#Find mean, median, min and max closing price of fund symbol AAAU
stock_2019_fs_mean = stock_2019_df['close'].mean()
stock_2019_fs_min = stock_2019_df['close'].min()
stock_2019_fs_max = stock_2019_df['close'].max()
stock_2019_fs_median = stock_2019_df['close'].median()

stock_2020_fs_mean = stock_2020_df['close'].mean()
stock_2020_fs_min = stock_2020_df['close'].min()
stock_2020_fs_max = stock_2020_df['close'].max()
stock_2020_fs_median = stock_2020_df['close'].median()

stock_2021_fs_mean = stock_2021_df['close'].mean()
stock_2021_fs_min = stock_2021_df['close'].min()
stock_2021_fs_max = stock_2021_df['close'].max()
stock_2021_fs_median = stock_2021_df['close'].median()
#Create a dataframe with those  summary stats

stock_2021_summary_AAAU = pd.DataFrame.from_dict({"Mean": [stock_2019_fs_mean,stock_2020_fs_mean,stock_2021_fs_mean],
                                                "Median":[stock_2019_fs_median,stock_2020_fs_median,stock_2021_fs_median],
                                                "Minimum":[stock_2019_fs_min,stock_2020_fs_min, stock_2021_fs_min],
                                                "Maximum":[stock_2019_fs_max,stock_2020_fs_max,stock_2021_fs_max]}, orient = 'index')
stock_2021_summary_AAAU.rename(columns={0:"2019",1:"2020",2:"2021"})
```

```python
#Find mean, median, min and max closing price of the entire stock market
stock_2019_mean = stock_all_2019_df['close'].mean()
stock_2019_min = stock_all_2019_df['close'].min()
stock_2019_max = stock_all_2019_df['close'].max()
stock_2019_median = stock_all_2019_df['close'].median()

stock_2020_mean = stock_all_2020_df['close'].mean()
stock_2020_min = stock_all_2020_df['close'].min()
stock_2020_max = stock_all_2020_df['close'].max()
stock_2020_median = stock_all_2020_df['close'].median()

stock_2021_mean = stock_all_2021_df['close'].mean()
stock_2021_min = stock_all_2021_df['close'].min()
stock_2021_max = stock_all_2021_df['close'].max()
stock_2021_median = stock_all_2021_df['close'].median()
#Create a dataframe with those  summary stats

stock_2021_summary = pd.DataFrame.from_dict({"Mean": [stock_2019_mean,stock_2020_mean,stock_2021_mean],
                                                "Median":[stock_2019_median,stock_2020_median,stock_2021_median],
                                                "Minimum":[stock_2019_min,stock_2020_min, stock_2021_min],
                                                "Maximum":[stock_2019_max,stock_2020_max,stock_2021_max]}, orient = 'index')
stock_2021_summary.rename(columns={0:"2019",1:"2020",2:"2021"})
```
